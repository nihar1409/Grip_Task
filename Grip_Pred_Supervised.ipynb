```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
import plotly.express as px
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
```

**Data**

```python
import pandas as pd

data = {
    'Hours': [2.5, 5.1, 3.2, 8.5, 3.5, 1.5, 9.2, 5.5, 8.3, 2.7, 7.7, 5.9, 4.5, 3.3, 1.1, 8.9, 2.5, 1.9, 6.1, 7.4, 2.7, 4.8, 3.8, 6.9, 7.8],
    'Scores': [21, 47, 27, 75, 30, 20, 88, 60, 81, 25, 85, 62, 41, 42, 17, 95, 30, 24, 67, 69, 30, 54, 35, 76, 86]
}

study_hours_scores = pd.DataFrame(data)
study_hours_scores
```

**Plotting To Understand The Relation Of Columns**

```python
px.scatter(study_hours_scores, x="Hours", y="Scores", trendline = "ols")
```

**MAking Prediction Model**

```python
X = study_hours_scores[["Hours"]]
y = study_hours_scores[["Scores"]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=5)
```

```python
model = LinearRegression()
```

```python
lr= model.fit(X_train, y_train)
```

```python
pred = lr.predict(X_test)
```

```python
mse = mean_squared_error(y_test, pred)
np.sqrt(mse)
```

```python
score = lr.predict([[9.25]])
round(float(score),3)
```

```python
cvs = cross_val_score(lr, X, y, cv=2)
```

```python
round(np.mean(cvs),3)
```
